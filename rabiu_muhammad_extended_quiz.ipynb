```python
import numpy as np
import matplotlib.pyplot as plt
import copy
import seaborn as sns
```

```python
def ThreeDVar(B,H,R,y,xb):
    K=(B@H.T)@(np.linalg.inv(H@B@H.T+R))
    xhat=xb+K@(y-H@xb)
    return xhat
```

```python
R=0.25*np.eye(12)
Y=np.array([80,90,110,160,80,20,10,40,70,100,110,80])
xb=np.array([76.9,91,114.2,154.2,88.1,18.6,11.4,31.1,69.6,105.7,112.7,77.4])
H=np.eye(12)
B=0.1*np.eye(12)
```

```python
xhat=ThreeDVar(B,H,R,Y,xb)
```

```python
plt.plot(xhat,'o',label='Optimal estimate')
plt.plot(xb,'*',label='Background')
plt.plot(Y,'o',label='Data')
plt.legend()
plt.show()
```

We can observe from the above plot that the optimal estimates are very close to the background and the data, except in few situations like in April,May,August and october in which the optimal estimates are closer to the background than the data.

```python
R1=0.1*np.eye(12)
Y1=np.array([80,90,110,160,80,20,10,40,70,100,110,80])
xb1=np.array([76.9,91,114.2,154.2,88.1,18.6,11.4,31.1,69.6,105.7,112.7,77.4])
H1=np.eye(12)
B1=0.25*np.eye(12)
xhat1=ThreeDVar(B1,H1,R1,Y1,xb1)
plt.plot(xhat1,'o',label='Optimal estimate')
plt.plot(xb1,'*',label='Background')
plt.plot(Y1,'o',label='Data')
plt.legend()
plt.show()
```

In this case, the observation error variance is smaller than the background error variance, this leads to some situations where the background points deviate from the optimal and data points. As such the optimal estimates fit closer to the data than the background.
